```python
total_wins = stats.groupby('Team') ['Wins'].sum().reset_index()  #This creates a variable to store a calculation that adds up the number of wins and
                                                                  #groups the result by team
total_wins = total_wins.sort_values(by='Wins', ascending= False) #This sorts the total number of wins in descending order
total_wins                                                        #This calls the variable in order to display the result
```

```python
total_losses = stats.groupby(['Team', 'Season']) ['Losses'].sum().reset_index()  #This creates a variable to store a calculation that adds up the                                                                                  removes any potential index created by default and then sorts the result in descending order                                                                               #groups the result by team                 #This sorts the total number of wins in descending order                                                                         #This calls the variable in order to display the result
total_losses = total_losses.sort_values(by= 'Losses', ascending=False)   #number of losses per team and removes any potential indexes created by default
total_losses
```

```python
top5 = stats.groupby('Season').apply(lambda x:x.nlargest(5,'Wins')).reset_index(drop=True)
top5
```

```python
stats.loc[stats.groupby('Season')['Losses'].idxmax()] #This retrieves all the stats and the index number of the worst performing team on the table and
                                                       #orders the result by the season and losses columns using the groupby, loc and index max function
                       

```

```python
most_goals = stats.loc[stats.groupby('Season') ['Goals For'].idxmax()] #This retrieves all the stats and the index number of the best performing team and
most_goals                        #orders the result by the season and number of goals scored columns using the groupby, loc and index max function
                       
```

```python
plt.bar(most_goals['Season'],most_goals['Goals For']) #This visualizes the season which recorded the most number of goals
plt.xlabel('Season') # This places the season in the x axis
plt.ylabel('Total Goals') # This places the number of goals in the x axis
plt.title('Total Goals Per Season') # This titles the chart 'Total Goals Per Season'
plt.xticks(rotation=45) #This places the season figures in the x axis at a 45 degree angle
plt.show() #This removes the plot function description that is displayed by default after a chart is created
```

```python
leastgoals=stats.loc[stats.groupby('Season')['Goals For'].idxmin()] #This retrieves all the stats and the index number of the worst performing team and
 leastgoals                          #orders the result by the season and number of goals scored columns using the groupby, loc and index max function                      

```

```python
plt.bar(leastgoals['Season'],leastgoals['Ties']) #This visualizes the season which recorded the most draws and ties per game
plt.xlabel('Season')  #This places the season in the x axis
plt.ylabel('Draws/Ties')  # This places the number of draws/ties in the x axis
plt.title('Season with Most Draws/Ties') # This titles the chart 'Total Goals Per Season'
plt.rotation = 45 
```

```python
StrongestTeam=stats.loc[stats.groupby('Season') ['Goals Against'].idxmin().reset_index(drop=True)]  #This retrieves all the stats and the index number of 
StrongestTeam                                  # the strongest team and orders the result by the season and number of goals against them scored columns using the groupby, loc and index max function
                       
```

```python
for team in StrongestTeam['Team'].unique(): #This retreives all the unique teams in the team column to avoid duplicates
    team_data = StrongestTeam[StrongestTeam['Team']==team] #This stores all the teams retrieved in the 'team data' variable
    plt.plot(team_data['Season'],team_data['Goals Against'],marker='o',label=team) #This creates a plot chart to show the strongest teams
plt.xlabel('Team') #This places the team in the x axis
plt.ylabel('Goals Conceded') #This places the goals conceded in the y axis
plt.title('Strongest Team Per Season') #This titles the chart 'Strongest Team Per Season'
plt.xticks(rotation=45) #This places the season figures in the x axis at a 45 degree angle
plt.legend()#This creates a legend to help show the different teams
plt.show() #This removes the plot function description that is displayed by default after a chart is created
        
```
